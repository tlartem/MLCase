```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math as m
```

```python
tt1 = t1.iloc[:, 2:]
tt1.head()
```

```python
correlation = tt1.corr()
sns.heatmap(correlation, annot=True)
plt.show()
```

```python
a = t1['feature_1'].corr(t1['feature_2'].apply(np.log))
print(a)
```

```python
plt.figure(figsize = (12, 6))
sns.scatterplot(data=t1, x = 'feature_1', y = 'feature_2')
plt.show()
```

```python
correlations = []
d = []
for i in range(1,419):
  for j in range(1,419):
    c = t1['feature_'+str(i)].corr(t1['feature_'+str(j)].apply(np.log))
    if (abs(c) > 0.8) and (i!=j):
      print(t1['feature_'+str(i)].corr(t1['feature_'+str(j)].apply(np.log)), i, j)
      d.append((i,j))
      correlations.append(c)
print(d)
```

```python
for i in d:
  plt.figure(figsize = (12, 6))
  sns.scatterplot(data=t1, x = 'feature_'+str(i[0]), y = 'feature_'+str(i[1]))
  plt.show()
  print(t1['feature_'+str(i[0])].corr(t1['feature_'+str(i[1])].apply(np.log)))
  print()
```

```python
covariations = []
for i in d:
  co = np.cov(t1['feature_'+str(i[0])], t1['feature_'+str(i[1])])[0, 1]
  print(co)
  covariations.append(co)
```

```python
print(covariations)
print(correlations)
```

```python
corr = ['feature_'+str(i[0]) for i in d ]
corr += ['feature_'+str(i[1]) for i in d]
print(corr)
```

```python
features = list(set(corr))
print(len(features))
```

```python

sns.pairplot(t1[features], diag_kind='kde')
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
G = nx.Graph()
for i in features:
    G.add_node(i)
```

```python
for i in d:
    feature_1 = t1['feature_' + str(i[0])].replace([np.inf, -np.inf], np.nan).dropna()
    feature_2 = t1['feature_' + str(i[1])].replace([np.inf, -np.inf], np.nan).dropna()

    merged_data = pd.concat([feature_1, feature_2], axis=1).dropna()

    if len(merged_data) > 1:
        corr, p_value = pearsonr(merged_data.iloc[:, 0], merged_data.iloc[:, 1])
        if abs(corr) > 0.8 and p_value < 0.05:
            G.add_edge('feature_' + str(i[0]), 'feature_' + str(i[1]), weight=abs(corr))
```

```python
plt.figure(figsize=(12, 12))
pos = nx.spring_layout(G, seed=42)
nx.draw_networkx_nodes(G, pos, node_size=50, node_color='skyblue')
nx.draw_networkx_edges(G, pos, width=1, alpha=0.7, edge_color='gray')
nx.draw_networkx_labels(G, pos, font_size=8, font_color='darkblue')
# edge_labels = nx.get_edge_attributes(G, 'weight')
# nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=6)
plt.title("Network Graph of Significant Covariations")
plt.show()
```
